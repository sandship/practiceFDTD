```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.01 # x方向空間差分間隔[m]
dy = 0.01 # y方向空間差分間隔[m]
dz = 0.01 # z方向空間差分間隔[m]

c = 2.99792458e8 # 光速[m/s]

dt = 0.99/(c * np.sqrt((1.0/dx ** 2 + 1.0/dy ** 2 + 1.0/dz ** 2))) # 時間差分間隔[s] c.f.Courantの安定条件
freq = 1.0e9 # 周波数[Hz]

nx = 100 # x方向計算点数
ny = 100 # y方向計算点数
nz = 100 # z方向計算点数

nt = 2 ** 10 # 計算ステップ数
nf = 2 ** 10
```

```python
tl = np.arange(0, nt*dt, dt)
fl = np.linspace(0, 1/dt, nf)
```

```python
f = np.sin(2 * np.pi * freq * tl)
plt.plot(tl, f)
```

```python
F = np.fft.fft(f)
amp = np.abs(F)
plt.plot(fl[:100], amp[:100])
```
